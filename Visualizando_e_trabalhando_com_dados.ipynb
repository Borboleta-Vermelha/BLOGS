```python
import pandas as pd
```

```python
base = pd.read_csv('Iris.csv')
```

```python
base
```

```python
base = base.drop(columns='Id')
```

```python
base.columns = ['SÃ©pala_Comprimento', 'SÃ©pala_Largura', 'PÃ©tala_Comprimento', 'PÃ©tala_Largura', 'EspÃ©cies']
```

```python
base
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.unique(base['EspÃ©cies'])
```

```python
sns.countplot(base['EspÃ©cies']);
```

```python
valores = base.columns[:-1]
fig, ax = plt.subplots(len(valores), 1, figsize=(15, 15))
for i in range(len(valores)) :
    sns.boxplot(x=valores[i], data=base, ax=ax[i])
```

```python
dados = base.corr()
sns.heatmap(dados, annot=True, fmt='.2f', linewidths=0.5, linecolor='white', mask=np.triu(dados), cmap='Blues')
plt.show()
```

```python
sns.pairplot(base, hue='EspÃ©cies', corner=True, palette='Blues')
plt.show()
```
